### State variables distributions

```python
import gym
import matplotlib.pyplot as plt
```

```python
def random_games(n_games):
    env = gym.make("LunarLander-v2")
    x = []
    y = []
    vx = []
    vy = []
    theta = []
    omega = []
    for i in range(n_games):
        obs = env.reset()
        done = False
        while not done:
            action = env.action_space.sample()
            x.append(obs[0])
            y.append(obs[1])
            vx.append(obs[2])
            vy.append(obs[3])
            theta.append(obs[4])
            omega.append(obs[5])
            obs, reward, done, info = env.step(action)
    return x, y, vx, vy, theta, omega
```

```python
x_list, y_list, vx_list, vy_list, theta_list, omega_list = random_games(1000)
```

```python
print(f'max value:{max(x_list)}, min value: {min(x_list)}')
plt.hist(x_list, bins=50)
plt.title('x')
plt.show()
```

```python
print(f'max value:{max(y_list)}, min value: {min(y_list)}')
plt.hist(y_list, bins=50)
plt.title('y')
plt.show()
```

```python
print(f'max value:{max(vx_list)}, min value: {min(vx_list)}')
plt.hist(vx_list, bins=50)
plt.title('vx')
plt.show()
```

```python
print(f'max value:{max(vy_list)}, min value: {min(vy_list)}')
plt.hist(vy_list, bins=50)
plt.title('vy')
plt.show()
```

```python
print(f'max value:{max(theta_list)}, min value: {min(theta_list)}')
plt.hist(theta_list, bins=50)
plt.title('theta')
plt.show()
```

```python
print(f'max value:{max(omega_list)}, min value: {min(omega_list)}')
plt.hist(omega_list, bins=50)
plt.title('omega')
plt.show()
```
